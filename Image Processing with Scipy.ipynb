### Flip image

```python
from scipy import ndimage
import numpy as np
```

```python
flip = np.fliplr(img)
plt.imshow(flip,cmap='gray')
```

```python
flip = np.flipud(img)
plt.imshow(flip,cmap='gray')
```

### Rotate img

```python
rotated = ndimage.rotate(img,45)
plt.imshow(rotated,cmap='gray')
```

```python
rotated = ndimage.rotate(img,45,reshape=False)
plt.imshow(rotated,cmap='gray')
```

### Filters

```python
uniform_filter = ndimage.uniform_filter(img,size=3) # blurring filter
plt.imshow(uniform_filter,cmap='gray')
```

```python
uniform_filter = ndimage.uniform_filter(img,size=9) # blurring filter
plt.imshow(uniform_filter,cmap='gray')
```

### Gausian Filter

```python
gaussian_filter = ndimage.gaussian_filter(img,sigma=3) # blurring filter
plt.imshow(gaussian_filter,cmap='gray')
```

```python
gaussian_filter = ndimage.gaussian_filter(img,sigma=7) # blurring filter
plt.imshow(gaussian_filter,cmap='gray')
```

```python
median_filter = ndimage.median_filter(img,size=3) # blurring filter
plt.imshow(median_filter,cmap='gray')
```

### Edge Detection

```python
sobel_img = ndimage.sobel(img,)
plt.imshow(sobel_img,cmap='gray')
```

```python
sobel_img = ndimage.sobel(img,1)
plt.imshow(sobel_img,cmap='gray')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

```python
img = io.imread('./Rohit.jfif',as_gray=True)
plt.imshow(img,cmap='gray')
```

```python
from skimage.transform import rescale, resize, downscale_local_mean
```

### Rescale

```python
rescaled_img = rescale(img,1/4,anti_aliasing=False)
plt.imshow(rescaled_img,cmap='gray')
```

```python
img.shape,rescaled_img.shape
```

### Resize

```python
resized_img = resize(img,(200,200))
plt.imshow(resized_img,cmap='gray')
```

```python
resized_img.shape
```

### Edge Detection

```python
img.shape
```

```python
from skimage.filters import sobel, roberts, scharr, prewitt
```

```python
edges = sobel(img)
plt.imshow(edges,cmap='gray')
```

```python
edges = prewitt(img)
plt.imshow(edges,cmap='gray')
```

```python
edges = scharr(img)
plt.imshow(edges,cmap='gray')
```

```python
edges = roberts(img)
plt.imshow(edges,cmap='gray')
```

```python
from skimage.feature import canny
```

```python
edges_canny = canny(img,sigma=3)
plt.imshow(edges_canny,cmap='gray')
```
